```python
# ‡¶™‡ßç‡¶∞‡¶Ø‡¶º‡ßã‡¶ú‡¶®‡ßÄ‡¶Ø‡¶º ‡¶≤‡¶æ‡¶á‡¶¨‡ßç‡¶∞‡ßá‡¶∞‡¶ø ‡¶á‡¶Æ‡¶™‡ßã‡¶∞‡ßç‡¶ü
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. CSV ‡¶´‡¶æ‡¶á‡¶≤ ‡¶≤‡ßã‡¶°
df = pd.read_csv("sales_data.csv")
```

```python
# 2. ‡¶°‡ßá‡¶ü‡¶æ‡¶∞ ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡ß´ ‡¶≤‡¶æ‡¶á‡¶® ‡¶¶‡ßá‡¶ñ‡¶æ
print("üìå ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡ß´‡¶ü‡¶ø ‡¶∞‡ßã:")
print(df.head())
```

```python
# 3. Missing Values ‡¶ö‡ßá‡¶ï
print("\nüìå Missing Values:")
print(df.isnull().sum())
```

```python
# 4. Date ‡¶ï‡¶≤‡¶æ‡¶Æ‡¶ï‡ßá ‡¶§‡¶æ‡¶∞‡¶ø‡¶ñ‡ßá ‡¶ï‡¶®‡¶≠‡¶æ‡¶∞‡ßç‡¶ü ‡¶ï‡¶∞‡¶æ
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 5. ‡¶Æ‡ßã‡¶ü ‡¶∏‡ßá‡¶≤‡¶∏ ‡¶è‡¶¨‡¶Ç ‡¶ï‡ßã‡ßü‡¶æ‡¶®‡ßç‡¶ü‡¶ø‡¶ü‡¶ø
total_sales = df['Sales'].sum()
total_quantity = df['Quantity'].sum()
print(f"\nüí∞ ‡¶Æ‡ßã‡¶ü ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø: {total_sales}")
print(f"üì¶ ‡¶Æ‡ßã‡¶ü ‡¶™‡¶£‡ßç‡¶Ø‡ßá‡¶∞ ‡¶∏‡¶Ç‡¶ñ‡ßç‡¶Ø‡¶æ: {total_quantity}")
```

```python
# 6. Region ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶Æ‡ßã‡¶ü ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
print("\nüìå Region ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶Æ‡ßã‡¶ü ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø:")
print(region_sales)
```

```python
# 7. Category ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶ó‡ßú ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø
category_avg_sales = df.groupby('Category')['Sales'].mean().reset_index()
print("\nüìå Category ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶ó‡ßú ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø:")
print(category_avg_sales)
```

```python
# 8. Visualization - Region Sales Bar Chart
plt.figure(figsize=(6,4))
sns.barplot(data=region_sales, x='Region', y='Sales', palette='viridis')
plt.title("Region ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶Æ‡ßã‡¶ü ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.show()
```

```python
# 9. Visualization - Category Pie Chart
plt.figure(figsize=(5,5))
plt.pie(category_avg_sales['Sales'], labels=category_avg_sales['Category'], autopct='%1.1f%%')
plt.title("Category ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶ó‡ßú ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø‡¶∞ ‡¶π‡¶æ‡¶∞")
plt.show()
```

```python
# 10. ‡¶Æ‡¶æ‡¶∏ ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø‡¶∞ ‡¶ü‡ßç‡¶∞‡ßá‡¶®‡ßç‡¶°
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(7,4))
# Convert the 'Month' column to string type for plotting
monthly_sales['Month'] = monthly_sales['Month'].astype(str)
sns.lineplot(data=monthly_sales, x='Month', y='Sales', marker='o')
plt.title("‡¶Æ‡¶æ‡¶∏ ‡¶Ö‡¶®‡ßÅ‡¶Ø‡¶æ‡ßü‡ßÄ ‡¶¨‡¶ø‡¶ï‡ßç‡¶∞‡¶ø‡¶∞ ‡¶ü‡ßç‡¶∞‡ßá‡¶®‡ßç‡¶°")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```
